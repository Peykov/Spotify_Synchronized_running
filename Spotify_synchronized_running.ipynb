# Spotify Synchronized Running

### Synch your music BPM with current running/cycling pace
___
___

#### **Project description**:
##### An application for wearable (GARMIN) devices that syncs your music with your running/cycling pace.

### Scope: 

### 1. Develop a simple Connect IQ App sending data to a server using a Bluetooth connection

-  IQ SDK Application built in Monkey C language

### 2.  Setting a Flask server 

- Seting up a local server that listens to incoming pace data from the device.
    - cmd: $ ipconfig (ip address) 
- Device: Fenix 5s Plus

### 3. Processing

- capturing the incoming data in the Flask route, storing, manipulation and sending instructions to the Spotify API
- a playlist is loaded before the run and sorted by BPM in bins. The program needs to navigate through the bins and play the songs in the correct order.

### Spotify side

- sorting the music playlist by BPM in groups of simililar songs for the various running paces/stages
- when a new song/songs are imported to the playlist a pipline is triggered and the song is added to the correct cluster

### UI/UX. Buttons of the watch application

    - Next bin (+ 5 BPM) 
    - Previous bin (- 5 BPM)
    - Next song in the current bin
    - Auto: Sync with current running cadence
    - Other metrics: Heart rate, pace
